#Number Series

```python
start = 0
stop = 10
step = 1

# Built-in Python Range
print range(stop)
print range(start, stop)
print range(start, stop, step)
print

# Numpy arange does the same thing but makes numpy arrays
print np.arange(stop)
print np.arange(start, stop)
print np.arange(start, stop, step)
print

# Generate series with fixed number of elements equally spaced
# Here both start and stop are included in the elements
print np.linspace(start, stop, 10) # 10 elements
print np.linspace(start, stop, 5) # 5 elements
print

# Generate series in logrithmic space
# This means that elements will be denser at the beginning and sparser near the end
print np.logspace(start, stop, 10)
print np.logspace(start, stop, 20)
```

```python
# Pull random numbers from the distributions
size = 10
mean = 1
sd = 1
var = sd**2 # Variance is standard deviation squared

# Normal
print np.random.randn(size)*sd + mean

# Poisson
print np.random.poisson(mean, size)

# Binomial
# binomial(n, p) where n is the number of trials and p is the probability of each one being successful.
# The mean = n*p and the variance = mean(1 - p)
print np.random.binomial(2*mean, .5, size)
```

#Regression with Statsmodels

```python
import statsmodels.formula.api as smf
from statsmodels.graphics.api import abline_plot
```

```python
x = np.arange(10)

# Generate y with some random noise
y = x*2 + np.random.randn(10)*2

# Take a look at it
figure, ax = pl.subplots(figsize=(10,10))
ax.scatter(x, y)
```

```python
# Create a dictionary so we can use formula.ols
d = {'x': x, 'y': y}

# Run regression and print summary
mod = smf.ols(formula='y ~ x', data = d).fit()
print mod.summary()
```

```python
# Plot points again
figure, ax = pl.subplots(figsize=(10,10))
ax.scatter(x, y)

# Plot regression line
abline_plot(model_results=mod, color='red', ax=ax)

ax.set_title('Linear Regression Line')
```

```python
# Polynomial Regression

# Create x, x**2 and y
x = np.arange(50)
x2 = x**2

# y = 2*x^2 + 5*x + noise
y = x2*2 + x*5 + np.random.randn(50)*10

# Store in a dict
d = {'x': x, 'x2': x2, 'y': y}

# Run regression against x2 and x
mod = smf.ols(formula='y ~ x2 + x', data = d).fit()
print mod.summary()

# Plot points again
figure, ax = pl.subplots(figsize=(10,10))
ax.scatter(x, y, s=100)

# Plot regression
y2 = mod.params[0] + x2*mod.params[1] + x*mod.params[2]
ax.plot(x, y2, 'r--')

ax.set_title('Polynomial Regression Line')
```

```python
# Let's see how much better this is than using a linear regression model

# Run the regression just on x
mod2 = smf.ols(formula='y ~ x', data = d).fit()

# Graph them just for clarity
figure, ax = pl.subplots(figsize=(10,10))
ax.scatter(x, y, s=100)

# Plot polynomial regression
y2 = mod.params[0] + x2*mod.params[1] + x*mod.params[2]
ax.plot(x, y2, 'r--')

# Plot linear regression
y2 = mod2.params[0] + x*mod2.params[1]
ax.plot(x, y2, 'g--')

ax.set_title('Polynomial vs Linear')

# Compare the two tests by passing the simpler model into the other's compare_lr_test function
# This returns (chisquare test result, p-value, difference in degrees of freedom)
print mod.compare_lr_test(mod2)
```

# Fitting a distribution

```python
import scipy

# Generate some data poisson distributed with mean=5
data = np.random.poisson(5, 5000)

# Get estimates for mean an sd fitting a _normal_ to it
mean, sd = scipy.stats.norm.fit(data)

# Generate a pdf so we can plot it
x = np.linspace(-5,20,100)
y = scipy.stats.norm.pdf(x, mean, sd)

figure, ax = pl.subplots(figsize=(10,10))

# Plot with normed=True so that the probabilities are normalized so they add up to 0
ax.hist(data, normed=True, bins=10)

# Plot the fitted pdf
ax.plot(x, y, 'r--')
```

```python
# Let's see how good of a fit that is
# Refer to HW4/goodness_of_fit.ipynb for expected values as the fit gets better.
# Generally lower is better, but it's hard to say when it's "good"

# KS Test
print scipy.stats.kstest(data, 'norm', args=(mean,sd))

# Anderson
print scipy.stats.anderson(data, 'norm')

# Define a function to run KL Divergence test
def kl(dist, m, s):
    num_bins=20
    hist, bin_edges = np.histogram(dist, bins=num_bins)
    bin_centers = bin_edges[:-1] + (bin_edges[1] - bin_edges[0]) * .5
    norm = scipy.stats.norm.pdf(bin_centers, loc=m, scale=s)
    return scipy.stats.entropy(hist, norm)

print kl(data, mean, sd)
```

```python
# Reading a CSV

# If we had a local file, i.e. citibike.csv, we could just simple use pd.read_csv('citibike.csv')
# Instead we can get it directly from the internet

import pandas as pd
from zipfile import ZipFile
from StringIO import StringIO
import urllib2

# Create a function to pull citibike data from the web directly into memory for a given month.
# month can be any of the following: ['01', '02', '03', '04', '05', '06']
def getCitibikeData(month):
    # Create the URL to the citibike zip file
    url = "https://s3.amazonaws.com/tripdata/2015%s-citibike-tripdata.zip" % month

    # Download the file into 'r'
    r = urllib2.urlopen(url).read()

    # Open the zip into 'f'
    f = ZipFile(StringIO(r))

    # Open the csv inside the zip
    csv = f.open("2015%s-citibike-tripdata.csv" % month)

    # Finally, return the dataframe
    return pd.read_csv(csv)

# Get a dataframe with data from january 2015
data = getCitibikeData('01')
```

```python
# Take a look at the data

# Print a list of columns
print data.columns

# Print stats for the numeric columns
data.describe()
```

```python
# Get a single column

# Access either with dictionary syntax
print data['tripduration'].head()

# Or with attribute syntax
print data.tripduration.head()

# This won't work with fields that have spaces in their names, i.e. 'start station id'
print data['start station id'].head()

# .head() just gets the first few entries in a numpy array.
```

```python
# We could rename those columns if we wanted to remove weird characters such as spaces
# Pass a dictionary with current : replace
# So we just want to replace spaces with underscores here
columns = {
    'start station id': 'start_station_id',
    'end station id': 'end_station_id'
}

# data.rename returns a new copy of data so assign it back to data to save the change.
data = data.rename(columns=columns)

# Now we can access data.start_station_id
print data.start_station_id.head()
```

```python
# We can create new columns in the dataframe with derived data, for example age.
data['age'] = 2015 - data['birth year']

print data['age'].head()
```

```python
# We can filter the data if we only want a subset.
# The first step is creating a "mask" array.  A mask array is an array of boolean values, True and False.
# Wherever there is a True we want to get the value and False we don't.

# For example, creating a mask for age > 45
# Compare the output to the above cell.
mask = (data['age'] > 45)
print mask.head()
```

```python
# Then the mask can be applied to the dataframe to get those rows which correspond to the True values.
print data[mask].head()
```

```python
# Multiple filters can be combined with and (&), or (|), etc.
# Create a mask where the age is greater than 45 and the gender is 1
mask = (data['age'] > 45) & (data['gender'] == 1)
print mask.head()
print data[mask].head()
```
